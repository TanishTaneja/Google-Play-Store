# Numeric Type Conversion: Examine the Number of Installs

**Challenge**: How many apps had over 1 billion (that's right - BILLION) installations? How many apps just had a single install? 

Check the datatype of the Installs column.

Count the number of apps at each level of installations. 

Convert the number of installations (the Installs column) to a numeric data type. Hint: this is a 2-step process. You'll have make sure you remove non-numeric characters first. 

```python
df_apps_clean['Installs'] = df_apps_clean['Installs'].str.replace('(\D+)', '')
df_apps_clean['Installs']=pd.to_numeric(df_apps_clean["Installs"])
```

```python
# billion_install=df_apps_clean[df_apps_clean["Installs"]>=1000000000].Installs.count()
# single_install=df_apps_clean[df_apps_clean["Installs"]==1].Installs.count()
# print(f"Apps with billion install {billion_install}")
# print(f"apps with single install {single_install}")
df_apps_clean[["App","Installs"]].groupby("Installs").count()
# df_apps_clean.info()
```

# Find the Most Expensive Apps, Filter out the Junk, and Calculate a (ballpark) Sales Revenue Estimate

Let's examine the Price column more closely.

**Challenge**: Convert the price column to numeric data. Then investigate the top 20 most expensive apps in the dataset.

Remove all apps that cost more than $250 from the `df_apps_clean` DataFrame.

Add a column called 'Revenue_Estimate' to the DataFrame. This column should hold the price of the app times the number of installs. What are the top 10 highest grossing paid apps according to this estimate? Out of the top 10 highest grossing paid apps, how many are games?


```python
df_apps_clean["Price"]=df_apps_clean["Price"].astype(str).str.replace('$', '')
df_apps_clean['Price']=pd.to_numeric(df_apps_clean["Price"])
df_apps_clean.sort_values("Price",ascending=False).head(20)
```

### The most expensive apps sub $250

```python
df_apps_clean.drop(df_apps_clean[df_apps_clean["Price"]>250].index,inplace=True)
df_apps_clean.sort_values("Price",ascending=False).head()
```

### Highest Grossing Paid Apps (ballpark estimate)

```python
df_apps_clean["Revenue_Estimate"]=df_apps_clean["Price"]*df_apps_clean["Installs"]
df_apps_clean.sort_values("Revenue_Estimate",ascending=False).head(10)
```

# Plotly Bar Charts & Scatter Plots: Analysing App Categories

```python
# top10_category = df_apps_clean.Category.value_counts()[:10]
top10_category=df_apps_clean.groupby('Category').agg({'App': pd.Series.count})
top10_category=top10_category.sort_values("App",ascending=False)
top10_category.head()
```

### Vertical Bar Chart - Highest Competition (Number of Apps)

```python
bar=px.bar(x=top10_category.head(10).index,y=top10_category.head(10).App,title="Apps Per Category")
bar.update_layout(xaxis_title="Category",yaxis_title="No. Of Apps",yaxis=dict(
    tickvals=list(range(0,2001,200))
))
bar.show()
```

### Horizontal Bar Chart - Most Popular Categories (Highest Downloads)

```python
# category_installs = df_apps_clean.groupby('Category').agg({'Installs': pd.Series.sum})
category_installs=df_apps_clean[["Category","Installs"]].groupby("Category").sum()
category_installs.sort_values(by="Installs",ascending=True,inplace=True)
```

```python
bar=px.bar(x=category_installs.Installs,y=category_installs.index,title="Category Popularity")
bar.update_layout(xaxis_title="Installs",yaxis_title="Category")
bar.show()
```

```python
cat_merged_df = pd.merge(top10_category, category_installs, on='Category', how="inner")
cat_merged_df.sort_values("Installs",ascending=False,inplace=True)
cat_merged_df.head()
```

```python
sc=px.scatter(cat_merged_df,x="App",y="Installs",color="Installs",hover_name=cat_merged_df.index,title="Category concentration",size="App")
sc.update_layout(yaxis=dict(
  type='log'  
))
sc.show()
```

## Summary
<i><strong>This analysis suggests that in categories like Family, Tools, and Games, there are numerous apps competing for downloads, with no single app dominating the market. This indicates a higher level of competition and diversity within these categories, as users have a wide range of options to choose from.

On the other hand, in categories like Video Players and Entertainment, a few apps capture the majority of downloads, indicating a concentration of market share among these dominant apps. This suggests that the competition in these categories may be more intense, with a smaller number of apps controlling a larger portion of the user base.</strong></i>

# Grouped Bar Charts: Free vs. Paid Apps per Category

```python
df_apps_clean.Type.value_counts()
```

```python
# df_apps_clean[["Category","Type","App"]].groupby(["Category","Type"],as_index=False).count()
df_free_vs_paid=df_apps_clean.groupby(["Category","Type"],as_index=False).agg({
    "App":pd.Series.count
})
df_free_vs_paid
```

```python
bar=px.bar(
        df_free_vs_paid,
        x="Category",
        y="App",
        title="Free vs Paid app by category",
        color="Type",
        barmode="group",
        height=600
    )
bar.update_layout(
    xaxis_title="Category",
    yaxis=dict(
        type='log',
        title="Number of Apps",
    ),
    font_size=15
)
bar.update_xaxes(
    categoryorder="total descending",
    tickangle=90
)
bar.show()
```

```python
box=px.box(df_apps_clean,y="Installs",x="Type",color="Type",notched=True,points="all",height=600)
box.update_layout(
    yaxis=dict(
        type='log',
    ),
    font_size=15
    
)
box.show()
```

```python
box=px.box(df_apps_clean[df_apps_clean["Type"]=="Paid"],y="Revenue_Estimate",x="Category")
box.update_layout(
    yaxis=dict(
        type='log',
    ), 
    xaxis={'categoryorder':'min ascending'}
)
box.show()
```
